```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sn
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.options.display.max_colwidth = 200
pd.read_excel('Data_Description.xlsx', sheet_name='Description')[['Variable', 'Definition']]
```

```python
df = pd.read_csv('sample_data_intw.csv')
df.head()
df.shape
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2,random_state=1)
"train:", train.shape
"test:", test.shape
```

```python
train.drop(['Unnamed: 0','msisdn','pcircle','pdate'],axis=1,inplace=True)
test.drop(['Unnamed: 0','msisdn','pcircle','pdate'],axis=1,inplace=True)
```

```python
train.fillna(train.mean(), inplace=True)
test.fillna(test.mean(), inplace=True)
```

```python
# check
train.apply(lambda x: pd.to_numeric(x, errors='coerce'))
test.apply(lambda x: pd.to_numeric(x, errors='coerce'))
```

```python
from sklearn.utils import resample
train_positive=train[train.label==1]
train_negative=train[train.label==0]

print("train:", train.shape, "= positive:", train_positive.shape, "+ negative:", train_negative.shape)

train_upsamp=resample(train_negative,replace=True,n_samples=train_positive.shape[0],random_state=1)
train_upsamp=pd.concat([train_positive,train_upsamp])

print("upsampled train:", train_upsamp.shape, "= positives:", train_upsamp[train_upsamp['label']==1].shape, "+ upsampled negatives:", train_upsamp[train_upsamp['label']==0].shape)
```

```python
y_train = train_upsamp.label
y_test = test.label
```

```python
cols = [str(i) for i in train.columns]
```

```python
from sklearn.preprocessing import Normalizer
train_normalized = Normalizer().fit_transform(train_upsamp)
train_normalized = pd.DataFrame(train_normalized, columns=cols)
test_normalized = Normalizer().fit_transform(test)
test_normalized = pd.DataFrame(test_normalized, columns=cols)
```

```python
x_train = train_normalized.copy().drop(['label'],axis=1,inplace=True)
x_test = test_normalized.copy().drop(['label'],axis=1,inplace=True)
```

```python
x_train.shape
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score
from sklearn.metrics import classification_report
```

```python
def train_and_predict(x_train, y_train, model) :
    model.fit(x_train, y_train)
    prediction = model.predict(x_test)
    
    accuracy = round(accuracy_score(prediction,y_test)*100,2)
    precision = round(precision_score(prediction,y_test)*100,2)
    recall = round(recall_score(prediction,y_test)*100,2)
    
    print("accuracy:", accuracy)
    print("precision:", precision)
    print("recall:", recall)
    
    print(classification_report(prediction,y_test))
```
